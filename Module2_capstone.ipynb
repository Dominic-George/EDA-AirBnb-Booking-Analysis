#### Chart - 1

````python
# Chart - 1 visualization code
#First calculate the averege price by neighbourhood_group
#performing the groupby operation for neighbourhood_group and price columns
nh_group_avg_price=df.groupby('neighbourhood_group')['price'].mean()
#sorting in descending order
nh_group_avg_price_sorted=nh_group_avg_price.sort_values(ascending=False)
print(nh_group_avg_price_sorted)
````

````python
# Plotting the figure
plt.figure(figsize=(10, 6))
# Plotting the bar chart
nh_group_avg_price_sorted.plot(kind='bar', color='grey')

# Adding titles and labels
plt.title('Average Price by Neighbourhood Group',fontsize=16)
plt.xlabel('Neighbourhood Group',fontsize=15,color='indigo')
plt.ylabel('Price',fontsize=15,color='indigo')
plt.xticks(rotation=360)
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Adding labels on top of each bar
for index, value in enumerate(nh_group_avg_price_sorted):
    plt.text(index, value + 1, round(value, 2), ha='center', va='bottom', fontsize=9)
plt.tight_layout()
# Display the plot
plt.show()
````

The reason I chose this specific chart is that vertical bar charts are the most effective way to depict or illustrate comparisons between groups or categories.If it is further arranged in ascending order, it will be simpler for the observer to determine the outcome or gain the insight, which is the highest percentage in relation to the local community.

````python
# Chart - 2 visualization code
#performing the groupby operations for host_name,neighbourhood_group and finding which host has more bookings or counts.We are using neighbourhood_group since there are simlilar names in different neighbourhood groups.
listings_count = df.groupby(['host_name','neighbourhood_group'])['calculated_host_listings_count'].max().reset_index()
#sorting the values in descending order
sorted_listings_count = listings_count.sort_values(by='calculated_host_listings_count', ascending=False)
sorted_listings_count.head(10)
````
